# Part 2: Parameter Derivation

```python
# convert data to monthly returns
monthly_prices = df_monthly_prices.values
# number of monthly returns for a stock
n_months = df_monthly_prices.shape[0] - 1
# number of stocks
n_assets = df_monthly_prices.shape[1]
monthly_returns = np.zeros((n_months,n_assets))
for i in range(n_months):
    monthly_returns[i,:] = (monthly_prices[i+1,:] / monthly_prices[i,:]) - 1

# calculate mean returns
r_mean = np.mean(monthly_returns,axis = 0)
mu = r_mean

# calculate return covariance matrix
r_cov_matrix = np.zeros((n_assets,n_assets))
for i in range(n_assets):
    for j in range(n_assets):
        r_cov_matrix[i,j] = np.sum((monthly_returns[:,i] - r_mean[i]) * (monthly_returns[:,j] - r_mean[j])) / (n_months-1)
Q = r_cov_matrix

# return variances
r_variance = np.diag(r_cov_matrix)
```

```python
# calculate risk-free rate
# data is from U.S. Department of The Treasury:
# https://www.treasury.gov/resource-center/data-chart-center/interest-rates/Pages/TextView.aspx?data=yieldYear&year=2008
Sep_2008_30YTBond_yield = np.array([4.43,4.36,4.32,4.27,4.27,4.26,4.20,4.32,4.12,4.08,4.08,4.14,4.36,4.41,4.43,4.40,4.40,4.36,4.13,4.31])
r_f_yearly = (1 + np.mean(Sep_2008_30YTBond_yield) / 100) - 1
r_f_monthly = (1 + np.mean(Sep_2008_30YTBond_yield) / 100) ** (1/12) - 1
```

```python
# data of outstanding shares on September 30th in billions (from Macrotrends website)
outstanding_shares = [2.279,0.625,1.948,1.136,4.658,1.593,3.95,0.583,3.331,3.445,25.260,1.384,6.736,2.831,5.178,0.711,0.274,5.893,2.845,0.455]

# closing price on September 30th
stock_price_daily_not_adj = pd.DataFrame({"F": F["Close"],
                             "CAT": CAT["Close"],
                             "DIS": DIS["Close"],
                             "MCD": MCD["Close"],
                             "KO": KO["Close"],
                             "PEP": PEP["Close"],
                             "WMT": WMT["Close"],
                             "C": C["Close"],
                             "WFC": WFC["Close"],
                             "JPM": JPM["Close"],
                             "AAPL": AAPL["Close"],
                             "IBM": IBM["Close"],
                             "PFE": PFE["Close"],
                             "JNJ": JNJ["Close"],
                             "XOM": XOM["Close"],
                             "MRO": MRO["Close"],
                             "ED": ED["Close"],
                             "T": T["Close"],
                             "VZ": VZ["Close"],
                             "NEM": NEM["Close"]}) 
september_30_close_price = stock_price_daily_not_adj.iloc[-1].values

# weights of market portfolio
w_mkt = (outstanding_shares * september_30_close_price) / np.sum(outstanding_shares * september_30_close_price)

# expected return of market portfolio, 0.007857237697971167
r_expected_mkt = np.dot(mu,w_mkt)

# return variance of market portfolio, 0.0008387844405966018
r_var_mkt = np.dot(np.dot(Q,w_mkt),w_mkt)

# risk aversion coefficient, 5.194036425263101
coef_lambda = (r_expected_mkt - r_f_monthly) / r_var_mkt
```

```python
# compute distancing coefficients delta for 95% confidence level
epsilon_95 = -norm.ppf(0.025)
delta_95 = epsilon_95 * ((r_variance) ** 0.5) / (n_months ** 0.5)

# compute distancing coefficients delta for 90% confidence level
epsilon_90 = -norm.ppf(0.05)
delta_90 = epsilon_90 * ((r_variance) ** 0.5) / (n_months ** 0.5)
```

# Part 3: Compute MVO, Robust MVO, Risk Parity Portfolios

```python
import cvxpy as cvp
import ipopt
```

```python
# MVO portfolio
c = -mu
e = np.ones(n_assets)
w = cvp.Variable(20)
prob = cvp.Problem(cvp.Minimize((1/2) * cvp.quad_form(w,Q * 2 * coef_lambda) + c.T @ w),[e.T @ w == 1])
prob.solve(solver=cvp.CPLEX,verbose = False, cplex_params = {"qpmethod":6})
w_MVO = w.value
```

```python
# robust MVO portfolio with 95% confidence level
e = np.ones(n_assets)
w = cvp.Variable(20)
y = cvp.Variable(20)
prob = cvp.Problem(cvp.Minimize((1/2) * cvp.quad_form(w,Q * 2 * coef_lambda) - mu.T @ w + delta_95.T @ y),[e.T @ w == 1,y >= w, y >= -w])
prob.solve(solver=cvp.CPLEX,verbose = False, cplex_params = {"qpmethod":6})
w_robust_MVO_95 = w.value
```

```python
# robust MVO portfolio with 90% confidence level
e = np.ones(n_assets)
w = cvp.Variable(20)
y = cvp.Variable(20)
prob = cvp.Problem(cvp.Minimize((1/2) * cvp.quad_form(w,Q * 2 * coef_lambda) - mu.T @ w + delta_90.T @ y),[e.T @ w == 1,y >= w, y >= -w])
prob.solve(solver=cvp.CPLEX,verbose = False, cplex_params = {"qpmethod":6})
w_robust_MVO_90 = w.value
```

```python
# risk parity portfolio
class erc(object):
    def __init__(self):
        pass

    def objective(self, x):
        # The callback for calculating the objective
        y = x[:-1] * np.dot(Q, x[:-1])
        fval = 0
        for i in range(n_assets):
            fval += (y[i] - x[-1]) * (y[i] - x[-1])
        return fval

    def gradient(self, x):
        # The callback for calculating the gradient
        grad = np.zeros(n_assets + 1)
        # Insert your gradient computations here
        # You can use finite differences to check the gradient
        y = x[:-1] * np.dot(Q, x[:-1])
        for i in range(n_assets):
            for grad_i in range(n_assets):
                    if i == grad_i:
                        grad[grad_i] += 2 * (y[i] - x[-1]) * (y[i] + x[i] * Q[i,i])
                    else:
                        grad[grad_i] += 2 * (y[i] - x[-1]) * (x[i] * Q[i,grad_i])
            grad[n_assets] += 2 * (y[i] - x[-1]) * (-1)
        return grad

    def constraints(self, x):
        # The callback for calculating the constraints
        return ([1.0] * n_assets + [0])
    
    def jacobian(self, x):
        # The callback for calculating the Jacobian
        return np.array([[1.0] * n_assets + [0]])

    def intermediate(
            self,
            alg_mod,
            iter_count,
            obj_value,
            inf_pr,
            inf_du,
            mu,
            d_norm,
            regularization_size,
            alpha_du,
            alpha_pr,
            ls_trials
            ):
        # Example for the use of the intermediate callback. uncomment next line for print
        # print("Objective value at iteration #%d is - %g" % (iter_count, obj_value))
        return

lb = [0.0] * n_assets + [-np.inf] # lower bounds on variables
ub = [1.0] * n_assets + [np.inf] # upper bounds on variables
cl = [1]        # lower bounds on constraints
cu = [1]        # upper bounds on constraints
w0 = [1.0/n_assets] * n_assets + [0]

# Define IPOPT problem
nlp = ipopt.problem(n=len(w0), m=len(cl), problem_obj=erc(), lb=lb, ub=ub, cl=cl, cu=cu)
     
# Set the IPOPT options
nlp.addOption('jac_c_constant'.encode('utf-8'), 'yes'.encode('utf-8'))
nlp.addOption('hessian_approximation'.encode('utf-8'), 'limited-memory'.encode('utf-8'))
nlp.addOption('mu_strategy'.encode('utf-8'), 'adaptive'.encode('utf-8'))
nlp.addOption('tol'.encode('utf-8'), 1e-30)

# Solve the problem
weights, info = nlp.solve(w0)
w_risk_parity = weights[:-1]
```

# Part 6: Efficient Frontiers

```python
# estimated and actual MVO frontier
lambda_space = np.linspace(0.1,coef_lambda * 5 ,500)
est_w_front_MVO = []
est_std_front_MVO = []
est_ret_front_MVO = []
act_std_front_MVO = []
act_ret_front_MVO = []
c = -mu_est
Q = Q_est
e = np.ones(n_assets)
for lambda_i in lambda_space: 
    w = cvp.Variable(20)
    prob = cvp.Problem(cvp.Minimize((1/2) * cvp.quad_form(w,Q * 2 * lambda_i) + c.T @ w),[e.T @ w == 1])
    prob.solve(solver=cvp.CPLEX,verbose = False, cplex_params = {"qpmethod":6})
    weights = w.value
    est_w_front_MVO.append(weights)
    est_std_front_MVO.append((weights.T @ Q_est @ weights) ** 0.5)
    est_ret_front_MVO.append(mu_est.T @ weights)
    act_std_front_MVO.append((weights.T @ Q_true_Oct @ weights) ** 0.5)
    act_ret_front_MVO.append(mu_true_Oct.T @ weights)

# true MVO frontier
lambda_space = np.linspace(1,1000,200)
true_w_front_MVO = []
true_std_front_MVO = []
true_ret_front_MVO = []
c = -mu_true_Oct
Q = Q_true_Oct
e = np.ones(n_assets)
for lambda_i in lambda_space: 
    w = cvp.Variable(20)
    prob = cvp.Problem(cvp.Minimize((1/2) * cvp.quad_form(w,Q * 2 * lambda_i) + c.T @ w),[e.T @ w == 1])
    prob.solve(solver=cvp.CPLEX,verbose = False, cplex_params = {"qpmethod":6})
    weights = w.value
    true_w_front_MVO.append(weights)
    true_std_front_MVO.append((weights.T @ Q_true_Oct @ weights) ** 0.5)
    true_ret_front_MVO.append(mu_true_Oct.T @ weights)
    
# estimated and actual 95% robust MVO frontier
lambda_space = np.linspace(0.01,coef_lambda * 5,500)
est_w_front_robust_95 = []
est_std_front_robust_95 = []
est_ret_front_robust_95 = []
act_std_front_robust_95 = []
act_ret_front_robust_95 = []
mu = mu_est
Q = Q_est
e = np.ones(n_assets)
for lambda_i in lambda_space: 
    w = cvp.Variable(20)
    y = cvp.Variable(20)
    prob = cvp.Problem(cvp.Minimize((1/2) * cvp.quad_form(w,Q * 2 * lambda_i) - mu.T @ w + delta_95.T @ y),[e.T @ w == 1,y >= w, y >= -w])
    prob.solve(solver=cvp.CPLEX,verbose = False, cplex_params = {"qpmethod":6})
    weights = w.value
    est_w_front_robust_95.append(weights)
    est_std_front_robust_95.append((weights.T @ Q_est @ weights) ** 0.5)
    est_ret_front_robust_95.append(mu_est.T @ weights)
    act_std_front_robust_95.append((weights.T @ Q_true_Oct @ weights) ** 0.5)
    act_ret_front_robust_95.append(mu_true_Oct.T @ weights)
    
# estimated and actual 90% robust MVO frontier
lambda_space = np.linspace(0.01,coef_lambda * 5,500)
est_w_front_robust_90 = []
est_std_front_robust_90 = []
est_ret_front_robust_90 = []
act_std_front_robust_90 = []
act_ret_front_robust_90 = []
mu = mu_est
Q = Q_est
e = np.ones(n_assets)
for lambda_i in lambda_space: 
    w = cvp.Variable(20)
    y = cvp.Variable(20)
    prob = cvp.Problem(cvp.Minimize((1/2) * cvp.quad_form(w,Q * 2 * lambda_i) - mu.T @ w + delta_90.T @ y),[e.T @ w == 1,y >= w, y >= -w])
    prob.solve(solver=cvp.CPLEX,verbose = False, cplex_params = {"qpmethod":6})
    weights = w.value
    est_w_front_robust_90.append(weights)
    est_std_front_robust_90.append((weights.T @ Q_est @ weights) ** 0.5)
    est_ret_front_robust_90.append(mu_est.T @ weights)
    act_std_front_robust_90.append((weights.T @ Q_true_Oct @ weights) ** 0.5)
    act_ret_front_robust_90.append(mu_true_Oct.T @ weights)
```
